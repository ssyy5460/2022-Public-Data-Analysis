# 평균값으로 나누기

```python
mean_1 = df['어린이집수'].mean()
mean_2 = df['국공립/민간비율'].mean()
mean_3 = df['보육교사수'].mean()
mean_4 = df['신혼부부'].mean()


df.head()
```

## 가중치 x

```python
total_li = []
for row in df.values.tolist():
    gu, ch_count, na_no, tea_n, tea_per, total, birth_rate, merr, chi = row[0:]
    
    ch_count = (ch_count/mean_1) 
    na_no = (na_no/mean_2)
    total = (total/df['전체 충족률'].mean())
    chi = (chi/df['보육시설 아동비율'].mean()) 
    tea_per = (tea_per/df['보육교사 1인당 아동수'].mean()) 
    tea_n = (tea_n/ mean_3) 
    birth_rate = (birth_rate/df['일반출생률'].mean()) 
    merr = (merr  / mean_4) 
    
    total_li.append([gu, ch_count, na_no, tea_n, tea_per, total, birth_rate, merr, chi ])
```

```python
ab = pd.DataFrame(total_li)
ab.columns = ['지역', '어린이집수', '국공립/민간비율','보육교사수', '보육교사 1인당 아동수', '전체 충족률','일반출생률','신혼부부','보육시설 아동비율']
ab['총합'] = ab['어린이집수'] + ab['국공립/민간비율'] + ab['보육교사 1인당 아동수'] + ab['보육교사수'] +ab['전체 충족률'] + ab['일반출생률']+ ab['신혼부부'] + ab['보육시설 아동비율'] 

ab['순위'] = ab['총합'].rank(ascending=False)
ab.sort_values(by='순위').head()
```

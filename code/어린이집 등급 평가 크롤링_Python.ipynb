```python
import time
```

```python
from selenium import webdriver

browser = webdriver.Chrome()
browser.get("http://info.childcare.go.kr/info/cera/search/findNurseryGrade.jsp")
```

```python
# 시/도 : 서울특별시 입력
browser.find_element("xpath", '//*[@id="sidoCode"]').click()
browser.implicitly_wait(3)
browser.find_element("xpath","//select/option[@value='11000']").click()
```

```python
total_name = []
total_state = []
total_grade = []
```

```python
for k in range(2,27):
    # 시/군/구 : 2 ~ 26까지 입력
    browser.find_element("xpath", '//*[@id="sigunguCode"]').click()
    browser.implicitly_wait(3)
    browser.find_element("xpath",'//*[@id="sigunguCode"]/option[%d]' %k).click()
    time.sleep(2)
    
    # 검색 클릭
    browser.find_element("xpath",'//*[@id="listFrm"]/div/div[2]/div/div[3]/a/img').click()
    time.sleep(2)
    
    # 정보공시 클릭
    browser.find_element("xpath",'//*[@id="pageLoadSort"]/ul[2]/li[2]/a').click()
    time.sleep(2)
    
    # 1페이지 크롤링
    for i in range(1,11):
        name = browser.find_element("xpath", '//*[@id="pageLoadResult"]/div/table/tbody/tr[%d]/td[3]/a' %i).text
        state = browser.find_element("xpath", '//*[@id="pageLoadResult"]/div/table/tbody/tr[%d]/td[4]' %i).text
        grade = browser.find_element("xpath", '//*[@id="pageLoadResult"]/div/table/tbody/tr[%d]/td[6]' %i).text

        total_name.append(name)
        total_state.append(state)
        total_grade.append(grade)
        
    # 2페이지 이후 페이지 크롤링
    try :
        for j in range(2,11):
            browser.find_element("xpath", '//a[@href="#page_%d"]' %j).click()
            time.sleep(3)
            for h in range(1,11):
                name = browser.find_element("xpath", '//*[@id="pageLoadResult"]/div/table/tbody/tr[%d]/td[3]/a' %h).text
                state = browser.find_element("xpath", '//*[@id="pageLoadResult"]/div/table/tbody/tr[%d]/td[4]' %h).text
                grade = browser.find_element("xpath", '//*[@id="pageLoadResult"]/div/table/tbody/tr[%d]/td[6]' %h).text

                total_name.append(name)
                total_state.append(state)
                total_grade.append(grade)
    
            
        # 다음 페이지 넘김
        try : 
            browser.find_element("xpath", '//a[@href="#page_next"]').click()
            time.sleep(3)
    
            # 11페이지 크롤링
            for i in range(1,11):
                name = browser.find_element("xpath", '//*[@id="pageLoadResult"]/div/table/tbody/tr[%d]/td[3]/a' %i).text
                state = browser.find_element("xpath", '//*[@id="pageLoadResult"]/div/table/tbody/tr[%d]/td[4]' %i).text
                grade = browser.find_element("xpath", '//*[@id="pageLoadResult"]/div/table/tbody/tr[%d]/td[6]' %i).text

                total_name.append(name)
                total_state.append(state)
                total_grade.append(grade)
        
            # 12페이지 이후 페이지 크롤링
            for j in range(12,21):
                browser.find_element("xpath", '//a[@href="#page_%d"]' %j).click()
                time.sleep(3)
                for i in range(1,11):
                    name = browser.find_element("xpath", '//*[@id="pageLoadResult"]/div/table/tbody/tr[%d]/td[3]/a' %i).text
                    state = browser.find_element("xpath", '//*[@id="pageLoadResult"]/div/table/tbody/tr[%d]/td[4]' %i).text
                    grade = browser.find_element("xpath", '//*[@id="pageLoadResult"]/div/table/tbody/tr[%d]/td[6]' %i).text
    
                    total_name.append(name)
                    total_state.append(state)
                    total_grade.append(grade)
            
            # 다음 페이지 넘김
            try : 
                browser.find_element("xpath", '//a[@href="#page_next"]').click()
                time.sleep(3)
    
                # 21페이지 크롤링
                for i in range(1,11):
                    name = browser.find_element("xpath", '//*[@id="pageLoadResult"]/div/table/tbody/tr[%d]/td[3]/a' %i).text
                    state = browser.find_element("xpath", '//*[@id="pageLoadResult"]/div/table/tbody/tr[%d]/td[4]' %i).text
                    grade = browser.find_element("xpath", '//*[@id="pageLoadResult"]/div/table/tbody/tr[%d]/td[6]' %i).text
    
                    total_name.append(name)
                    total_state.append(state)
                    total_grade.append(grade)
            
                # 22페이지 이후 페이지 크롤링
                for j in range(22,31):
                    browser.find_element("xpath", '//a[@href="#page_%d"]' %j).click()
                    time.sleep(3)
                    for i in range(1,11):
                        name = browser.find_element("xpath", '//*[@id="pageLoadResult"]/div/table/tbody/tr[%d]/td[3]/a' %i).text
                        state = browser.find_element("xpath", '//*[@id="pageLoadResult"]/div/table/tbody/tr[%d]/td[4]' %i).text
                        grade = browser.find_element("xpath", '//*[@id="pageLoadResult"]/div/table/tbody/tr[%d]/td[6]' %i).text
    
                        total_name.append(name)
                        total_state.append(state)
                        total_grade.append(grade)

                try :
                    browser.find_element("xpath", '//a[@href="#page_next"]').click()
                    time.sleep(3)
                        
                    # 31페이지 크롤링
                    for i in range(1,11):
                        name = browser.find_element("xpath", '//*[@id="pageLoadResult"]/div/table/tbody/tr[%d]/td[3]/a' %i).text
                        state = browser.find_element("xpath", '//*[@id="pageLoadResult"]/div/table/tbody/tr[%d]/td[4]' %i).text
                        grade = browser.find_element("xpath", '//*[@id="pageLoadResult"]/div/table/tbody/tr[%d]/td[6]' %i).text

                        total_name.append(name)
                        total_state.append(state)
                        total_grade.append(grade)
                    
                    # 32페이지 이후 페이지 크롤링
                    for j in range(32,40):
                        browser.find_element("xpath", '//a[@href="#page_%d"]' %j).click()
                        time.sleep(3)
            
                        for i in range(1,11):
                            name = browser.find_element("xpath", '//*[@id="pageLoadResult"]/div/table/tbody/tr[%d]/td[3]/a' %i).text
                            state = browser.find_element("xpath", '//*[@id="pageLoadResult"]/div/table/tbody/tr[%d]/td[4]' %i).text
                            grade = browser.find_element("xpath", '//*[@id="pageLoadResult"]/div/table/tbody/tr[%d]/td[6]' %i).text
    
                            total_name.append(name)
                            total_state.append(state)
                            total_grade.append(grade) 

                except :
                    print("error")
                
            except :
                print("error")
                
        except:
            print("error")
            
    except:
        print("error")
```

```python
import pandas as pd
data = pd.DataFrame(list(zip(total_name, total_state, total_grade)), columns = ["어린이집명", "시도/시군구", "평가결과"])
```

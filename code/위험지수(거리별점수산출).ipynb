```python
import pandas as pd
import numpy as np
```

```python
data_kids = pd.read_csv("./강서구 보행어린이 사고 지점별 거리 행렬.csv", encoding = "cp949")
```

```python
data_zone = pd.read_csv("./강서구 스쿨존 사고 지점별 거리 행렬.csv", encoding = "cp949")
```

```python
data_kids = data_kids.set_index('ID')
data_zone = data_zone.set_index('ID')
```

```python
data = pd.concat([data_zone, data_kids], axis = 1)
```

```python
data
```

```python
def get_score(s) : 
    if 0 <= s and s <= 100: 
        score = 3
    elif 100 < s and s <= 200 :
        score = 2
    elif 200 < s and s <= 300  : 
        score = 1 
    else : 
        score = 0
    return score
```

```python
columns = data.columns
```

```python
for i in range(0,18):
    data[columns[i]+"_score"] = data[columns[i]].apply(lambda x : get_score(x))
```

```python
data_score = data.iloc[:,18:]
data_score
```

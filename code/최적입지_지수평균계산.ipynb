```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../data/지수산출.csv", encoding = "cp949")
```

```python
data
```

```python
data_grouped = data.groupby(['CLUSTER_ID'])[["환경지수","접근지수","위험지수"]].mean().sort_values(by = ["위험지수"])
```

```python
data_grouped
```

```python
cluster = pd.DataFrame(list(data_grouped.index), columns = ["ID"])
```

```python
# MinMax 스케일링
from sklearn.preprocessing import MinMaxScaler

minmax = MinMaxScaler()
scale_minmax = minmax.fit_transform(data_grouped)

scale_df = pd.DataFrame(scale_minmax, columns = ["환경지수","접근지수","위험지수"])
scale_df = pd.concat([cluster,scale_df],axis = 1)
scale_df = scale_df.set_index('ID')
```

```python
scale_df["위험지수"] = - scale_df["위험지수"]
```

```python
scale_df
```

```python
pd.DataFrame(scale_df.sum(axis = 1).sort_values(ascending = False), columns = ["Min-Max_총괄지수"])
```

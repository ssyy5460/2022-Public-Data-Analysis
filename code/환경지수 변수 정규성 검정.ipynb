```python
import pandas as pd
import numpy as np
```

```python
import scipy.stats as stats
```

```python
data = pd.read_csv("./강서구 어린이집 환경치수 체험시설합침.csv", encoding = "cp949")
```

```python
data
```

```python
# shapiro 정규분포 검정
for i in range(9,16):
    test_stat, p_val = stats.shapiro(data.iloc[:,i])
    print(data.columns[i] + " Test-statistics : {}, p-value : {}".format(test_stat, p_val))
```
